```python
# Gráfico de Divisão de telas (Subplots)

# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt

# Carregando dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Girth com Volume
plt.scatter(base.Girth, base.Volume)
```

```python
# Girth com Height
plt.scatter(base.Girth, base.Height)
```

```python
# Height com Volume
plt.scatter(base.Height, base.Volume)
```

```python
# Histograma Volume
plt.hist(base.Volume)
```

```python
# Vamos colocar todos eles na mesma tela ("Dashboard")
plt.figure(1)

plt.subplot(2,2,1) # (Nº Linhas,Nº Colunas,Nº Posição de impressão)
plt.scatter(base.Girth, base.Volume)

plt.subplot(2,2,2)
plt.scatter(base.Girth, base.Height, marker = '*', color = 'red')
            
plt.subplot(2,2,3)
plt.scatter(base.Height, base.Volume, marker = '*')

plt.subplot(2,2,4)
plt.hist(base.Volume)
```
