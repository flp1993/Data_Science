```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axes3d
```

```python
'''
Dispersão você consegue ver 2 variáveis numéricas, uma
em cada eixo.

No gráfico de dispersão 3d você consegue adicionar
uma 3º variável numérica'''
```

```python
# Carregando Dataframe
base = pd.read_csv("orchard.csv")
base
```

```python
figura = plt.figure()
eixo = figura.add_subplot(1,1,1, projection = '3d')
eixo.scatter(base.decrease, base.rowpos, base.colpos)
eixo.set_xlabel('Decrease')
eixo.set_ylabel('Rowpos')
eixo.set_zlabel('Colpos')
```
