```python
# Gráfico de Densidade
# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn

# Carregando dados arvores
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Histograma com 6 divisões (bins) usando somente o 1º atributo
# Bins padrão = 10
plt.hist(base.iloc[:,1], bins = 6)
```

```python
# Agora vamos ver alguns exemplos de histograma com e sem Densidade
# kde é referente à densidade
```

```python
# Histograma ativo e Densidade desativada

srn.distplot(base.iloc[:,1], hist = True, kde = False, bins = 6,
            color = 'blue', hist_kws = {'edgecolor' : 'black'})
plt.title('Histograma Sofisticado')
```

```python
# Agora sem Histograma e com Densidade ativa

srn.distplot(base.iloc[:,1], hist = False, kde = True, bins = 6)
plt.title('Linha de Densidade')
```

```python
# Uma mesclagem das duas plotagens: Hist e Dens ativos
srn.distplot(base.iloc[:,1], hist = True, kde = True, bins = 6)
plt.title('Histograma com Linha de Densidade')
```
