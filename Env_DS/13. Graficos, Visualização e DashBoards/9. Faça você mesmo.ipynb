```python
# Imports
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
```

```python
# Carregando DF
dados = pd.read_csv('dados.csv', sep =';')
dados.head(), dados.shape
```

```python
grupo_PIB = dados['PIB']
grupo_MUNICIPIO = dados.groupby(['MUNICIPIO']).size()
grupo_VALOREMPENHO = dados['VALOREMPENHO']
grupo_PIB.describe(), grupo_MUNICIPIO, grupo_VALOREMPENHO.describe()
```

```python
srn.distplot(dados.iloc[:,2], hist = True, kde = True,
            color = 'gray',).set_title('PIB')
```

```python
srn.distplot(dados.iloc[:,3], hist = True, kde = True,
            color = 'gray',).set_title('Valor Empenho')
```

```python
srn.boxplot(dados.PIB, showfliers = False).set_title('PIB')
```

```python
srn.boxplot(dados.VALOREMPENHO, showfliers = False).set_title('Valor Empenho')
```

```python
plt.figure(1)

plt.subplot(2,2,1)
srn.distplot(dados.iloc[:,2], hist = True, kde = True,
            color = 'gray',).set_title('PIB')

plt.subplot(2,2,2)
srn.boxplot(dados.PIB, 
             showfliers = False).set_title('PIB')

plt.subplot(2,2,3)
srn.distplot(dados.iloc[:,3], hist = True, kde = True,
            color = 'gray',).set_title('Valor Empenho')

plt.subplot(2,2,4)
srn.boxplot(dados.VALOREMPENHO, 
            showfliers = False).set_title('Valor Empenho')

plt.tight_layout()
```

```python
top10_PIB = dados.sort_values('PIB').head(10)
top10_EMPENHO = dados.sort_values('VALOREMPENHO').head(10)
```

```python
top10_PIB.plot.bar(x = 'PIB', y = 'VALOREMPENHO',
                  color = 'gray')
```
