```python
# Gráfico de Dispersão

# Importando bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as srn

# Carregando dados
base = pd.read_csv('trees.csv')
base.head()
```

```python
# Muito usado para comparar 2 variáveis contínuas
```

```python
# Grafico de Dispersão considerando "Volume" e a "Circunferencia"
plt.scatter(base.Girth, base.Volume, color = 'blue')
plt.title('Árvores')
plt.ylabel('Circunferencia')
plt.xlabel('Volume')
```

```python
# Grafico de Linha
plt.plot(base.Girth, base.Volume)
plt.title('Árvores')
plt.ylabel('Circunferencia')
plt.xlabel('Volume')
```

```python
# Grafico de Dispersão com afastamento dos Dados (jitter)

# x_jitter é uma tremulação. Bastante usado para evitar 
# que os dados se sobreponham sobre os outros 
# no gráfico.

# fit_reg é a linha de tendencia

srn.regplot(base.Girth, base.Volume, data = base, x_jitter = 0.3,
           fit_reg = True)
```

```python
# Carregando Dataframe CO2
dfco2 = pd.read_csv('co2.csv')
dfco2.head()
```

```python
# Gráfico de dispersão, atr = conc e uptake, agrupado por type
srn.scatterplot(dfco2.conc, dfco2.uptake, hue = dfco2.Type)
```

```python
# Separando os registros de "Type"
q = dfco2.loc[dfco2['Type'] == 'Quebec']
m = dfco2.loc[dfco2['Type'] == 'Mississippi']
```

```python
# Gráfico de dispersão com vários plots
plt.figure()
plt.subplot(1,2,1)
srn.scatterplot(q.conc, q.uptake).set_title('Quebec')
plt.subplot(1,2,2)
srn.scatterplot(m.conc, m.uptake).set_title('Mississippi')
plt.tight_layout()
```

```python
ch = dfco2.loc[dfco2['Treatment'] == 'chilled'] 
nc = dfco2.loc[dfco2['Treatment'] == 'nonchilled']
```

```python
plt.figure()

plt.subplot(1,2,1)
srn.scatterplot(ch.conc, ch.uptake).set_title('chilled')

plt.subplot(1,2,2)
srn.scatterplot(nc.conc, nc.uptake).set_title('nonchilled')
```

```python
srn.catplot(x = 'conc', y = 'uptake', data = dfco2, jitter = False)
```
